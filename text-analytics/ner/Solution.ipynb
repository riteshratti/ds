# MODEL TRAINING 

```python
y_train_vectorized = keras.utils.to_categorical(y_train, 5)
```

```python

model = Sequential()

model.add(Embedding(3069, 1000, input_length=46))
# Recurrent layer
model.add(LSTM(46, input_shape=(46, 1000), return_sequences=True))

# Fully connected layer
model.add(Dense(5, activation='softmax'))


model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train, y_train_vectorized,batch_size = 100, epochs=10,verbose=1)
   
```
