```python
lines = itemnames 
```

```python
maxlen = 0
maxword = ''
sentences = []
for line in lines:
    #print(line.rstrip())
    words = []
    print(line)
    val = str(line).rstrip().lower()
    words = val.split(' ')
    newval = ''
    for i in words:
        i = i.replace('/','').replace(',','').replace('.','').replace('+','').replace('\'','').replace(' ','')
        newval = newval + i + ' '
        updatedString = newval[0:len(newval) -1 ]
    #val = re.sub('\W+','', val )
    newwords = []
    newwords.append(updatedString)
    sentences.append(newwords)
```

```python
all_words = sentences[:len(sentences)]
```

# Word 2 Vec Embeddings

```python
# Set values for various parameters
#uses skip gram on default ,good if you want features to know context of neighbourhood tokens
num_features = 300  # Word vector dimensionality                                              
num_workers = 4       # Number of threads to run in parallel
context = 2        # Context window size                                                                                    

# Initialize and train the model (this will take some time)
#print "Training model..."
model = word2vec.Word2Vec(workers=num_workers, vector_size=num_features,window = context, min_count=1, sample=0, sorted_vocab=0, hs=1, negative=0)
```

```python
sentences[1:10]
```

```python
all_words = sentences[:len(sentences)]
```

```python
model.build_vocab(all_words)
model.train(all_words,total_examples=model.corpus_count, epochs=10)
model_name = "item_name_sg_embedding_word2vec"    
model.save(model_name)
```
